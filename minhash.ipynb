```python
import pandas as pd
import numpy as np

def read_files(file_path):
    with open(file_path, 'r') as file:
        return file.read()

def create_shingles(text, k=5):
    words = text.split()
    return {''.join(words[i:i + k]) for i in range(len(words) - k + 1)}

file_paths = [
    'C:/Users/Nidhi/OneDrive/Desktop/text1.txt',
    'C:/Users/Nidhi/OneDrive/Desktop/text2.txt',
    'C:/Users/Nidhi/OneDrive/Desktop/text3.txt'
]

texts = [read_files(fp) for fp in file_paths]
shingles_list = [create_shingles(text, 4) for text in texts]

for i, shingles in enumerate(shingles_list):
    print(f"shingles {i+1}:\n" + '\n'.join(shingles) + "\n")

# Convert the universal set to a list
universal = list(set().union(*shingles_list))
incidence_matrix = pd.DataFrame(0, index=universal, columns=[f'Doc{i+1}' for i in range(len(shingles_list))])

for i, shingles in enumerate(shingles_list):
    for shingle in shingles:
        incidence_matrix.loc[shingle, f'Doc{i+1}'] = 1

print(incidence_matrix)

def hash_func(x, a, b):
    return (a * x + b) % 733

hashes = {
    'hash1': [hash_func(i, 2, 1) for i in range(len(universal))],
    'hash2': [hash_func(i, 3, 2) for i in range(len(universal))],
    'hash3': [hash_func(i, 5, 2) for i in range(len(universal))]
}

hash_df = pd.DataFrame(hashes, index=universal)
print(hash_df)

signature_matrix = pd.DataFrame(np.inf, index=hashes.keys(), columns=incidence_matrix.columns)

for col in incidence_matrix.columns:
    for row in incidence_matrix.index:
        if incidence_matrix.loc[row, col] == 1:
            for hash_function in signature_matrix.index:
                signature_matrix.loc[hash_function, col] = min(signature_matrix.loc[hash_function, col], hash_df.loc[row, hash_function])

print(signature_matrix)

def jaccard_similar(doc1, doc2):
    set1 = set(signature_matrix[doc1])
    set2 = set(signature_matrix[doc2])
    return len(set1.intersection(set2)) / len(set1.union(set2))

similar_doc1_2 = jaccard_similar('Doc1', 'Doc2')
similar_doc1_3 = jaccard_similar('Doc1', 'Doc3')
similar_doc2_3 = jaccard_similar('Doc2', 'Doc3')

print(f"Similarity between Doc1 and Doc2: {similar_doc1_2}")
print(f"Similarity between Doc1 and Doc3: {similar_doc1_3}")
print(f"Similarity between Doc2 and Doc3: {similar_doc2_3}")
```
